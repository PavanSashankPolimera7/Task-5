```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/BigMart Sales.csv')
```

```python
print(df)
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace( 'Low Fat','LF')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace( 'low fat','LF')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace( 'Regular','Reg')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace( 'reg','Reg')
```

```python
df.isnull().sum()
```

```python
df['Item_Weight'] = df['Item_Weight'].replace(np.nan,df['Item_Weight'].mean())
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
print(df)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

## **Pairplot**

```python
numeric_cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales','Item_Fat_Content']
sns.pairplot(df[numeric_cols],hue='Item_Fat_Content')
plt.show()
```


* Item_MRP vs Item_Outlet_Sales: A visible positive trend — higher MRP tends to result in higher sales.

* Item_Visibility and Item_Weight show no strong correlation with sales.

* Both Low Fat and Regular items are distributed similarly, with no major visual separation between them in most plots.



## **Heatmap**

```python
data = {
    'Item_Weight': [9.30, 5.92, 17.50, 19.20, 8.93],
    'Item_Visibility': [0.016, 0.019, 0.017, 0.000, 0.000],
    'Item_MRP': [249.81, 48.27, 141.62, 182.09, 53.86],
    'Item_Outlet_Sales': [3735.14, 443.42, 2097.27, 732.38, 994.71]
}
df2 = pd.DataFrame(data)
corr_matrix = df2.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

* Item_MRP vs Item_Outlet_Sales → Strong Positive Correlation
 Higher MRP values are closely associated with higher sales.
 Suggests pricing plays a significant role in sales performance.

* Item_Visibility and Item_Weight show weak or no correlation with other variables.
They do not significantly influence sales or pricing in this sample.

* Item_Weight vs Item_MRP / Sales
 Near-zero correlation
Indicates weight does not drive pricing or sales patterns.

## **Boxplot**

```python
sns.boxplot(x='Item_Type', y='Item_MRP', data=df)
plt.title('MRP by Item Type')
plt.xticks(rotation=90)
plt.show()
```

* There is a wide range of MRP values within most item types, indicating pricing diversity.

* Some categories like Household and Health and Hygiene tend to have higher median MRPs.

* Outliers are present in several categories, suggesting some items are priced significantly higher than others in the same group.

* A few categories like Starchy Foods or Breakfast have lower and narrower MRP ranges.




## **Scatterplot**

```python
sns.scatterplot(x='Item_MRP',y='Item_Outlet_Sales',hue='Item_Fat_Content',data=df)
plt.title('Item MRP vs Outle Sales')
plt.show()
```

* There is a clear upward trend: as Item MRP increases, Item Outlet Sales also tend to increase.

* This suggests a positive correlation between price and sales revenue.

* Both Low Fat and Regular items follow a similar distribution, showing no major impact of fat content on the price-sales relationship.

* A few high-sales outliers exist, indicating some expensive items perform exceptionally well.

## **Histogram**

```python
sns.histplot(data=df, x='Item_Outlet_Sales', bins=20,kde=False)
plt.title('Distribution of Item Outlet Sales')
plt.show()
```

* The majority of sales values are clustered towards the lower range, indicating many items have relatively low to moderate sales.

* There is a right-skewed distribution — a few items have very high sales acting as outliers.

* Most sales fall within the first few bins, showing a large number of low-selling items.

* The spread suggests sales vary widely across items.
